```python
# Lista com todos os pares de DataFrames para os gráficos
dataframes_pairs = [
    (df_pFinalRes_0_U_0_02_nonNewtonian, df_UzFinalRes_0_U_0_02_nonNewtonian, "U = 0.02 (nonNewtonian)", "blue", "cyan"),
    (df_pFinalRes_0_U_0_22_newtonianNoTurb, df_UzFinalRes_0_U_0_22_newtonianNoTurb, "U = 0.22 (newtonianNoTurb)", "red", "orangered"),
    (df_pFinalRes_0_U_0_385_nonNewtonianTurb, df_UzFinalRes_0_U_0_385_nonNewtonianTurb, "U = 0.385 (nonNewtonianTurb)", "green", "limegreen"),
    (df_pFinalRes_0_U_10_12_newtonianTurb, df_UzFinalRes_0_U_10_12_newtonianTurb, "U = 10.12 (newtonianTurb)", "purple", "violet")
]

# Determinar limites globais do eixo Y para o gráfico conjunto
global_y_min = min(min(df_p["Resíduo"].min(), df_uz["Resíduo"].min()) for df_p, df_uz, _, _, _ in dataframes_pairs)
global_y_max = max(max(df_p["Resíduo"].max(), df_uz["Resíduo"].max()) for df_p, df_uz, _, _, _ in dataframes_pairs)

# Gerar gráficos individuais
for df_p, df_uz, title, color_p, color_uz in dataframes_pairs:
    # Criar gráfico individual
    plt.figure(figsize=(10, 6))
    plt.plot(df_p["Tempo"], df_p["Resíduo"], label=f"p - {title}", color=color_p)
    plt.plot(df_uz["Tempo"], df_uz["Resíduo"], label=f"Uz - {title}", color=color_uz)
    plt.title(f"Resíduos - {title}")
    plt.xlabel("Tempo")
    plt.ylabel("Resíduo")

    # Definir escala logarítmica no eixo Y
    plt.yscale("log")

    # Definir ticks do eixo Y baseados em potências de 10
    y_ticks = np.logspace(np.floor(np.log10(global_y_min)), np.ceil(np.log10(global_y_max)), num=10)
    plt.yticks(y_ticks, [f"{ytick:.0e}" for ytick in y_ticks])

    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)  # Grade ajustada para a escala logarítmica
    plt.show()

# Criar gráfico com todos os dados juntos
plt.figure(figsize=(12, 8))

for df_p, df_uz, title, color_p, color_uz in dataframes_pairs:
    plt.plot(df_p["Tempo"], df_p["Resíduo"], label=f"p - {title}", color=color_p, linestyle='-')
    plt.plot(df_uz["Tempo"], df_uz["Resíduo"], label=f"Uz - {title}", color=color_uz, linestyle='--')

plt.title("Resíduos - Todos os Casos")
plt.xlabel("Tempo")
plt.ylabel("Resíduo")

# Definir escala logarítmica no eixo Y
plt.yscale("log")

# Definir ticks do eixo Y baseados em potências de 10
y_ticks = np.logspace(np.floor(np.log10(global_y_min)), np.ceil(np.log10(global_y_max)), num=10)
plt.yticks(y_ticks, [f"{ytick:.0e}" for ytick in y_ticks])

plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)  # Grade ajustada para a escala logarítmica
plt.show()
```

```python
# Gradientes de pressão
dp_dz_newtonian = 0.015519
dp_dz_nonnewtonian = 0.0001282

#Leitura dos dados de simulação (renomeados)
df_newtonian_sim_Re_Laminar = pd.read_csv('/content/drive/MyDrive/OpenFoam_Rodrigo_Gabriel/Caso1_Resultado_V3/U_0_22_newtonianNoTurb.csv')
df_nonnewtonian_sim_Re_Laminar = pd.read_csv('/content/drive/MyDrive/OpenFoam_Rodrigo_Gabriel/Caso1_Resultado_V3/U_0_02_nonNewtonian.csv')

# Cálculo dos perfis analíticos avaliados nos pontos simulados
# Para o fluido newtoniano:
df_newtonian_sim_Re_Laminar['U_analytical'] = (1 / (4 * nu)) * dp_dz_newtonian * (R**2 - df_newtonian_sim_Re_Laminar['Points_1']**2)

# Para o fluido não-newtoniano:
r_abs_nonnewt = np.abs(df_nonnewtonian_sim_Re_Laminar['Points_1'])
df_nonnewtonian_sim_Re_Laminar['U_analytical'] = (n / (n + 1)) * ((dp_dz_nonnewtonian / (2 * k))**(1 / n)) * \
    (R**((n + 1) / n)) * (1 - (r_abs_nonnewt / R)**((n + 1) / n))

# Criação de curvas analíticas contínuas (usando linspace)
# Definindo vetores radiais com base no intervalo dos dados simulados
r_newtonian_Re_Laminar = np.linspace(df_newtonian_sim_Re_Laminar['Points_1'].min(),
                                     df_newtonian_sim_Re_Laminar['Points_1'].max(), 200)
r_nonnewtonian_Re_Laminar = np.linspace(df_nonnewtonian_sim_Re_Laminar['Points_1'].min(),
                                        df_nonnewtonian_sim_Re_Laminar['Points_1'].max(), 200)

# Perfis analíticos contínuos:
u_newtonian_cont_Re_Laminar = (1 / (4 * nu)) * dp_dz_newtonian * (R**2 - r_newtonian_Re_Laminar**2)
u_nonnewtonian_cont_Re_Laminar = (n / (n + 1)) * ((dp_dz_nonnewtonian / (2 * k))**(1 / n)) * \
    (R**((n + 1) / n)) * (1 - (np.abs(r_nonnewtonian_Re_Laminar) / R)**((n + 1) / n))

# Cálculo do coeficiente R²
r2_newtonian_Re_Laminar = r2_score(df_newtonian_sim_Re_Laminar['U_2'], df_newtonian_sim_Re_Laminar['U_analytical'])
r2_nonnewtonian_Re_Laminar = r2_score(df_nonnewtonian_sim_Re_Laminar['U_2'], df_nonnewtonian_sim_Re_Laminar['U_analytical'])

# Gráfico para o fluido newtoniano
plt.figure(figsize=(10, 6))
plt.scatter(df_newtonian_sim_Re_Laminar['U_2'], df_newtonian_sim_Re_Laminar['Points_1'],
            label="Newtoniano Simulação U=0.22", color='blue', marker='o')
plt.plot(u_newtonian_cont_Re_Laminar, r_newtonian_Re_Laminar,
         label="Newtoniano Analítico", color='red', linewidth=2)
plt.title("Perfil de Velocidade - Fluido Newtoniano [MEL] - Re=100")
plt.xlabel("Velocidade no eixo Z do Mel (U_2)")
plt.ylabel("Posição radial r (Points_1)")
plt.legend()
plt.grid(True)
# Anotação central com R² (usando coordenadas do eixo: x=0.5, y=0.5)
plt.text(0.5, 0.5, f"$R^2$ = {r2_newtonian_Re_Laminar:.4f}", transform=plt.gca().transAxes,
         verticalalignment='center', horizontalalignment='center', fontsize=14, color='red')
plt.show()

# Gráfico para o fluido não-newtoniano
plt.figure(figsize=(10, 6))
plt.scatter(df_nonnewtonian_sim_Re_Laminar['U_2'], df_nonnewtonian_sim_Re_Laminar['Points_1'],
            label="Não Newtoniano Simulação U=0.02", color='green', marker='o')

plt.plot(u_nonnewtonian_cont_Re_Laminar, r_nonnewtonian_Re_Laminar,
         label="Não Newtoniano Analítico", color='darkorange', linewidth=2)
plt.title("Perfil de Velocidade - Fluido Não Newtoniano  [Lama de Perfuração] - Re=100")
plt.xlabel("Velocidade no eixo Z da Lama (U_2)")
plt.ylabel("Posição radial r (Points_1)")
plt.legend()
plt.grid(True)
plt.text(0.5, 0.5, f"$R^2$ = {r2_nonnewtonian_Re_Laminar:.4f}", transform=plt.gca().transAxes,
         verticalalignment='center', horizontalalignment='center', fontsize=14, color='darkorange')
plt.show()

# Gráfico combinado com ambos os casos
plt.figure(figsize=(10, 6))
# Newtoniano
plt.scatter(df_newtonian_sim_Re_Laminar['U_2'], df_newtonian_sim_Re_Laminar['Points_1'],
            label="Newtoniano Simulação U=0.22", color='blue', marker='o')
plt.plot(u_newtonian_cont_Re_Laminar, r_newtonian_Re_Laminar,
         label="Newtoniano Analítico", color='red', linewidth=2)
# Não-newtoniano
plt.scatter(df_nonnewtonian_sim_Re_Laminar['U_2'], df_nonnewtonian_sim_Re_Laminar['Points_1'],
            label="Não Newtoniano Simulação U=0.02", color='green', marker='o')
plt.plot(u_nonnewtonian_cont_Re_Laminar, r_nonnewtonian_Re_Laminar,
         label="Não Newtoniano Analítico", color='darkorange', linewidth=2)

plt.title("Comparação dos Perfis de Velocidade do Mel e da Lama de Perfuração - Re=100")
plt.xlabel("Velocidade no eixo Z (U_2)")
plt.ylabel("Posição radial r (Points_1)")
plt.legend()
plt.grid(True)

plt.text(0.3, 0.5, f"$R^2_{{Newtoniano}}$ = {r2_newtonian_Re_Laminar:.4f}", transform=plt.gca().transAxes,
         verticalalignment='center', horizontalalignment='center', fontsize=14, color='red')
plt.text(0.7, 0.5, f"$R^2_{{Não\,Newtoniano}}$ = {r2_nonnewtonian_Re_Laminar:.4f}", transform=plt.gca().transAxes,
         verticalalignment='center', horizontalalignment='center', fontsize=14, color='darkorange')
plt.show()
```

```python
# Gradientes de pressão
dp_dz_newtonian_Turbulento = 2.026
dp_dz_nonnewtonian_Turbulento = 0.00251

# Leitura dos dados de simulação (renomeados para caso turbulento)
df_newtonian_sim_Re_Turbulento = pd.read_csv('/content/drive/MyDrive/OpenFoam_Rodrigo_Gabriel/Caso1_Resultado_V3/U_10_12_newtonianTurb.csv')
df_nonnewtonian_sim_Re_Turbulento = pd.read_csv('/content/drive/MyDrive/OpenFoam_Rodrigo_Gabriel/Caso1_Resultado_V3/U_0_385_nonNewtonianTurb.csv')

# Cálculo dos perfis analíticos avaliados nos pontos simulados
# Para o fluido newtoniano:
df_newtonian_sim_Re_Turbulento['U_analytical'] = (1 / (4 * nu)) * dp_dz_newtonian_Turbulento * (R**2 - df_newtonian_sim_Re_Turbulento['Points_1']**2)

# Para o fluido não-newtoniano:
r_abs_nonnewt_Turbulento = np.abs(df_nonnewtonian_sim_Re_Turbulento['Points_1'])
df_nonnewtonian_sim_Re_Turbulento['U_analytical'] = (n / (n + 1)) * ((dp_dz_nonnewtonian_Turbulento / (2 * k))**(1 / n)) * \
    (R**((n + 1) / n)) * (1 - (r_abs_nonnewt_Turbulento / R)**((n + 1) / n))

#Criação de curvas analíticas contínuas (usando linspace)
# Definindo vetores radiais com base no intervalo dos dados simulados
r_newtonian_Re_Turbulento = np.linspace(df_newtonian_sim_Re_Turbulento['Points_1'].min(),
                                        df_newtonian_sim_Re_Turbulento['Points_1'].max(), 200)
r_nonnewtonian_Re_Turbulento = np.linspace(df_nonnewtonian_sim_Re_Turbulento['Points_1'].min(),
                                           df_nonnewtonian_sim_Re_Turbulento['Points_1'].max(), 200)

# Perfis analíticos contínuos:
u_newtonian_cont_Re_Turbulento = (1 / (4 * nu)) * dp_dz_newtonian_Turbulento * (R**2 - r_newtonian_Re_Turbulento**2)
u_nonnewtonian_cont_Re_Turbulento = (n / (n + 1)) * ((dp_dz_nonnewtonian_Turbulento / (2 * k))**(1 / n)) * \
    (R**((n + 1) / n)) * (1 - (np.abs(r_nonnewtonian_Re_Turbulento) / R)**((n + 1) / n))

#Cálculo do coeficiente R² para cada caso
r2_newtonian_Re_Turbulento = r2_score(df_newtonian_sim_Re_Turbulento['U_2'], df_newtonian_sim_Re_Turbulento['U_analytical'])
r2_nonnewtonian_Re_Turbulento = r2_score(df_nonnewtonian_sim_Re_Turbulento['U_2'], df_nonnewtonian_sim_Re_Turbulento['U_analytical'])

# Gráfico para o fluido newtoniano
plt.figure(figsize=(10, 6))
plt.scatter(df_newtonian_sim_Re_Turbulento['U_2'], df_newtonian_sim_Re_Turbulento['Points_1'],
            label="Newtoniano Simulação U=10.12", color='blue', marker='o')
plt.plot(u_newtonian_cont_Re_Turbulento, r_newtonian_Re_Turbulento,
         label="Newtoniano Analítico", color='red', linewidth=2)
plt.title("Perfil de Velocidade - Fluido Newtoniano [MEL] - Turbulento")
plt.xlabel("Velocidade no eixo Z do Mel (U_2)")
plt.ylabel("Posição radial r (Points_1)")
plt.legend()
plt.grid(True)
plt.text(0.5, 0.5, f"$R^2$ = {r2_newtonian_Re_Turbulento:.4f}", transform=plt.gca().transAxes,
         verticalalignment='center', horizontalalignment='center', fontsize=14, color='red')
plt.show()

#Gráfico para o fluido não-newtoniano
plt.figure(figsize=(10, 6))
plt.scatter(df_nonnewtonian_sim_Re_Turbulento['U_2'], df_nonnewtonian_sim_Re_Turbulento['Points_1'],
            label="Não Newtoniano Simulação U=0.385", color='green', marker='o')
plt.plot(u_nonnewtonian_cont_Re_Turbulento, r_nonnewtonian_Re_Turbulento,
         label="Não Newtoniano Analítico", color='darkorange', linewidth=2)
plt.title("Perfil de Velocidade - Fluido Não Newtoniano [Lama de Perfuração] - Turbulento")
plt.xlabel("Velocidade no eixo Z da Lama (U_2)")
plt.ylabel("Posição radial r (Points_1)")
plt.legend()
plt.grid(True)
plt.text(0.5, 0.5, f"$R^2$ = {r2_nonnewtonian_Re_Turbulento:.4f}", transform=plt.gca().transAxes,
         verticalalignment='center', horizontalalignment='center', fontsize=14, color='darkorange')
plt.show()

# Gráfico combinado com ambos os casos
plt.figure(figsize=(10, 6))
# Newtoniano
plt.scatter(df_newtonian_sim_Re_Turbulento['U_2'], df_newtonian_sim_Re_Turbulento['Points_1'],
            label="Newtoniano Simulação U=10.12", color='blue', marker='o')
plt.plot(u_newtonian_cont_Re_Turbulento, r_newtonian_Re_Turbulento,
         label="Newtoniano Analítico", color='red', linewidth=2)
# Não-newtoniano
plt.scatter(df_nonnewtonian_sim_Re_Turbulento['U_2'], df_nonnewtonian_sim_Re_Turbulento['Points_1'],
            label="Não Newtoniano Simulação U=0.385", color='green', marker='o')
plt.plot(u_nonnewtonian_cont_Re_Turbulento, r_nonnewtonian_Re_Turbulento,
         label="Não Newtoniano Analítico", color='darkorange', linewidth=2)

plt.title("Comparação dos Perfis de Velocidade do Mel e da Lama de Perfuração - Turbulento")
plt.xlabel("Velocidade no eixo Z (U_2)")
plt.ylabel("Posição radial r (Points_1)")
plt.legend()
plt.grid(True)
plt.text(0.3, 0.5, f"$R^2_{{Newtoniano}}$ = {r2_newtonian_Re_Turbulento:.4f}", transform=plt.gca().transAxes,
         verticalalignment='center', horizontalalignment='center', fontsize=14, color='red')
plt.text(0.7, 0.5, f"$R^2_{{Não\,Newtoniano}}$ = {r2_nonnewtonian_Re_Turbulento:.4f}", transform=plt.gca().transAxes,
         verticalalignment='center', horizontalalignment='center', fontsize=14, color='darkorange')
plt.show()
```
